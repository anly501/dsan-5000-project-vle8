```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Web-Scraping-Stats/cleaned_stats_data/player_stats_combined.csv", index_col=False)
df.rename(columns={'X3P':'ThreeP', 'X3PA': 'ThreeP_Attempted', 'X3P.': 'ThreeP_Percentage', 'FG.': 'FG_Percentage', 'FT.': 'FT_Percentage'}, inplace=True)
df['MP'] = df['MP'].str.split(':').str[0].astype(int)
df.head()
```

```python
Game_1980s = df[df['Year'].between(1980,1989)]
Game_1990s = df[df['Year'].between(1990,1999)]
Game_2000s = df[df['Year'].between(2000,2009)]
```

```python
#  QUESTION 1
print("Average Points Per Player in 1980s =", round(Game_1980s['PTS'].mean()), "points")
print("Average Points Per Player in 1990s =", round(Game_1990s['PTS'].mean()), "points")
print("Average Points Per Player in 2000s =", round(Game_2000s['PTS'].mean()), "points")
print("Overall Average Points Per Player =", round(df['PTS'].mean()), "points")
```

```python
# QUESTION 1 VISUAL
average_points_per_year = df.groupby(['Year'])['PTS'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='PTS', data=average_points_per_year, errorbar=None, palette='viridis')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.ylabel('Average Points per Player')
plt.title('Average Points per Player by Year')
plt.axhline(y=11, color='red', linestyle='--') # overall average
plt.savefig("../mysite/AvgPTS-Year.png")
plt.show()
```
